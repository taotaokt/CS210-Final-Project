```python
import pymysql

# MySQL connection configuration
DB_CONFIG = {
    'host': 'localhost',
    'user': 'root',
    'password': '123456',
    'local_infile': 1  # enable LOAD DATA LOCAL INFILE
}

def init_database():
    print("Connecting to MySQL...")
    conn = pymysql.connect(**DB_CONFIG)
    cursor = conn.cursor()
    
    print("Creating database if not exists: ecommerce")
    cursor.execute("CREATE DATABASE IF NOT EXISTS ecommerce;")
    cursor.execute("USE ecommerce;")

    # ----------------------------------------------------
    # Create fact table: fact_events
    # ----------------------------------------------------
    print("Recreating table: fact_events")
    cursor.execute("DROP TABLE IF EXISTS fact_events;")
    cursor.execute("""
    CREATE TABLE fact_events (
        event_time DATETIME,            -- timestamp of the user event
        event_type VARCHAR(20),         -- event type (e.g., view, cart, purchase)
        product_id INT,                 -- product ID
        category_id BIGINT,             -- category ID
        category_code VARCHAR(255),     -- category code
        brand VARCHAR(255),             -- product brand
        price DECIMAL(10, 2),           -- product price
        user_id INT,                    -- user ID
        user_session VARCHAR(100)       -- unique user session ID
    ) ENGINE=InnoDB;
    """)

    # ----------------------------------------------------
    # Create product dimension table: dim_products
    # ----------------------------------------------------
    print("Recreating table: dim_products")
    cursor.execute("DROP TABLE IF EXISTS dim_products;")
    cursor.execute("""
    CREATE TABLE dim_products (
        product_id INT PRIMARY KEY,     -- unique product identifier
        category_code VARCHAR(255),     -- product category
        brand VARCHAR(255),             -- brand name
        price DECIMAL(10, 2)            -- standard or listed price
    );
    """)

    # Create user dimension table: dim_users
    print("Recreating table: dim_users")
    cursor.execute("DROP TABLE IF EXISTS dim_users;")
    cursor.execute("""
    CREATE TABLE dim_users (
        user_id INT PRIMARY KEY,        -- unique user identifier
        first_active_time DATETIME,     -- first known activity timestamp
        last_active_time DATETIME,      -- most recent activity timestamp
        total_sessions INT              -- total number of sessions
    );
    """)

    conn.commit()
    conn.close()
    print("Database schema created successfully.")

if __name__ == "__main__":
    init_database()
```
